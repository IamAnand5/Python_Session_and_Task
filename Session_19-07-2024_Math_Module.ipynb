### Math Module:
- Mathematic function
sqrt(x), ceil(x), floor(x), fabs(x), log(x), sin(x), tan(x).......

- Random Module
 - functions to generate random numbers
 - Developing games, cryptography and to generate random numbers on the fly for authentication
 - random(): always generate some float value between 0 and 1 ( not inclusive)
 - randint(): generate random integer between two given numbers (inclusive)
 - uniform(): returns random float values between 2 given numbers(not inclusive)
 - randrange(start, stop, step): start is optional and default value is 0, step is optional and default value is 1
 - choice(): it won't return random numbers. It will return a random object from the given list or tuple

Packaging:
- It is an encapsulation mechanism to group related modules into a single unit.
- Package folder and directory which represents collection of python modules.
- Any folder or directory contains __init__.py file , is considered as a python package. 
- A package can contains sub packages also

OOPs:
- Class: In Python every thing is an object. To create objects we required some model, plan or blue print, is called a class
- We represent properties(attributes, variables) and actions(behavior, Method, functions) of object.
- Class contains both variables and methods
Syntax:

	methods: instance methods, static methods and class methods
